```python
import numpy as np
import matplotlib.pyplot as plt

# Create 100 numbers between 0 and 10
x = np.linspace(0, 10, 100)

# Calculate the sine of each x value
y = np.sin(x)

# Plot the sine wave
plt.plot(x, y, label='sin(x)', color='green')
plt.title("Sine Wave")
plt.xlabel("X Axis")
plt.ylabel("Sine Value")
plt.legend()  # Show label box
plt.show()
```

```python
import matplotlib.pyplot as plt

languages = ['Python', 'Java', 'C++', 'JavaScript']
popularity = [90, 75, 65, 80]

plt.bar(languages, popularity, color='orange')
plt.title("Programming Language Popularity")
plt.xlabel("Languages")
plt.ylabel("Popularity (%)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    'Language': ['Python', 'Java', 'C++', 'JavaScript'],
    'Users': [120, 95, 60, 110]
}
df = pd.DataFrame(data)

# Plot Bar Chart
plt.bar(df['Language'], df['Users'], color='skyblue')
plt.title("Number of Users by Language")
plt.xlabel("Programming Language")
plt.ylabel("Users (in millions)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random marks for 100 students
marks = np.random.randint(0, 100, 100)

# Plot histogram
plt.hist(marks, bins=10, color='purple', edgecolor='black')
plt.title("Student Marks Distribution")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 15, 20]
colors = ['gold', 'lightblue', 'lightcoral', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Language Usage Share")
plt.axis('equal')  # Makes it a perfect circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data: Hours studied vs. Marks scored
hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [40, 45, 50, 55, 60, 70, 75, 80]

plt.scatter(hours, marks, color='red')
plt.title("Hours Studied vs. Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.grid(True)
plt.show()
```
