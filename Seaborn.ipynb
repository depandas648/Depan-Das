```python
# Step 1: Check if seaborn is installed, else install it (only run in Jupyter or interactive)
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load a sample dataset from seaborn
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")

# Step 3: Advanced Chart 1 - Violin Plot (distribution + KDE)
plt.figure(figsize=(8, 5))
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot: Total Bill by Day")
plt.show()

# Step 4: Advanced Chart 2 - Heatmap (correlation matrix of iris dataset)
plt.figure(figsize=(8, 6))
corr = iris.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap: Iris Dataset Correlation")
plt.show()

# Step 5: Advanced Chart 3 - Pairplot (scatterplot matrix with categories)
sns.pairplot(iris, hue="species", height=2.5)
plt.suptitle("Pairplot: Iris Dataset", y=1.02)
plt.show()

# Step 6: Advanced Chart 4 - Jointplot (scatter + histograms)
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg", height=7)
plt.suptitle("Jointplot: Total Bill vs Tip with Regression", y=1.02)
plt.show()

# Step 7: Advanced Chart 5 - Swarm Plot (categorical scatter plot)
plt.figure(figsize=(8, 5))
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("Swarm Plot: Total Bill by Day and Gender")
plt.show()
```
