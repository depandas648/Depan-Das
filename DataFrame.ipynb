```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a simple table (DataFrame)
data = {
    'Year': [2020, 2021, 2022, 2023],
    'Sales': [100, 150, 130, 180]
}

df = pd.DataFrame(data)

# Plotting the data
plt.plot(df['Year'], df['Sales'], marker='o')  # x = Year, y = Sales
plt.title("Yearly Sales")  # Title on top
plt.xlabel("Year")         # Label below x-axis
plt.ylabel("Sales")        # Label next to y-axis
plt.grid(True)             # Adds a grid to the background
plt.show()                 # Displays the chart
```
